```python
#Q6write a python program that log an error message to a log file when a
# division by zero exception occurs
import logging
logging.basicConfig(filename ="errors.log", level = logging.ERROR, format="%(asctime)s-%(levelname)s-%(message)s")
def division(a, b):
    try:
        return a/b
    except ZeroDivisionError as e:
        logging.error("attempted division by zero. a=%s, b =%s", a, b)
        return None
result = division(4, 0)
print("result", result)
```

```python
#Q21 write a python program that read file and print the no of ocurance of a specific word
def counts(name, word):
    try:
        with open(name, 'r') as f:
            text = f.read()
        count = text.count(word)
        print(f"the word {word}found{count}")
    except FileNotFoundError:
        print("the file does not exist")
counts('files.txt', "apple")

```
